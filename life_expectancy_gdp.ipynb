```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

gdp = pd.read_csv('all_data.csv')

print(gdp.head())
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(data=gdp, x='Year', y='Life expectancy at birth (years)', hue='Country')
plt.title('Life Expectancy over Time')
plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth (years)')
plt.legend(title='Country')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(data=gdp, x='Year', y='GDP', hue='Country')
plt.title('GDP over Time')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.legend(title='Country')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(data=gdp, x='GDP', y='Life expectancy at birth (years)', hue='Country')
plt.title('Correlation between GDP and Life Expectancy')
plt.xlabel('GDP')
plt.ylabel('Life Expectancy at Birth (years)')
plt.legend(title='Country')
plt.show()

# Calculate correlation coefficient
correlation = gdp[['GDP', 'Life expectancy at birth (years)']].corr()
print("Correlation coefficient between GDP and Life Expectancy:\n", correlation)
```

```python
average_life_expectancy = gdp.groupby('Country')['Life expectancy at birth (years)'].mean()

print("Average Life Expectancy in each Country:")
print(average_life_expectancy)


# Plotting the average life expectancy for each country
plt.figure(figsize=(12, 6))
sns.barplot(x=average_life_expectancy.index, y=average_life_expectancy.values)
plt.title('Average Life Expectancy in Each Country')
plt.xlabel('Country')
plt.ylabel('Average Life Expectancy at Birth (years)')
plt.xticks(rotation=45)
plt.show()
```
